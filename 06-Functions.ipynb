# Functions

```python
def celsius_to_kelvin(temp): #temp is the argument
    return temp + 273.15
```

```python
print('Freezing point of water', celsius_to_kelvin(0))
print('Boiling point of water', celsius_to_kelvin(100))
```

```python
def fahr_to_celsius(temp):
    return (temp - 32) * (5/9)
```

```python
print('This is cold', fahr_to_celsius(0))
```

```python
def fahr_to_kelvin(temp):
    return celsius_to_kelvin(fahr_to_celsius(temp))
```

```python
def analyse(filename):
    data = numpy.loadtxt(fname=filename, delimiter=',')
    
    fig = plt.figure(figsize = (10,3))

    axes1 = fig.add_subplot(1,3,1)
    axes1.set_ylabel('average')
    axes1.plot(data.mean(axis=0))

    axes2 = fig.add_subplot(1,3,2)
    axes2.set_ylabel('max')
    axes2.plot(data.max(axis=0))

    axes3 = fig.add_subplot(1,3,3)
    axes3.set_ylabel('min')
    axes3.plot(data.min(axis=0))
    axes3.set_ylim(0,6)
    fig.tight_layout()
    plt.show()
```

```python
def detect_problems(filename):
    data = numpy.loadtxt(fname=filename, delimiter=',')
    
    if data.max(axis=0)[0] ==0 and data.max(axis=0)[20] == 20:
        print('Suspicious looking maxima')
    elif data.min(axis=0).sum() == 0:
        print('Minima add up to zero!')
    else:
        print('Seems OK')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy
import glob

filenames = sorted(glob.glob('data/inflammation*.csv'))

for file_i in filenames:
    print(file_i)
    analyse(file_i)
    detect_problems(file_i)
```
